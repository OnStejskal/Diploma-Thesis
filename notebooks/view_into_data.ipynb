```python
import torch
import matplotlib.pyplot as plt

# Assuming 'images' is a list of 3D tensors (channel, width, height)
# Example: images = [torch.rand(3, 100, 100), torch.rand(3, 100, 100)]

minsize = min(len(homogen), len(heterogen))
homogen = homogen[:minsize]
heterogen = heterogen[:minsize]
print(len(homogen), len(heterogen))

def non_zero_pixels(image):
    # Create a mask where all channels are non-zero
    mask = (image > 0).all(dim=0)
    # Apply the mask
    return image[:, mask]
def mean_intensity_values(filtered_image):
    # Calculate mean intensity for each channel
    return filtered_image.view(filtered_image.size(0), -1).mean(dim=1)

def standard_deviation_values(filtered_image):
    # Calculate standard deviation for each channel
    return filtered_image.view(filtered_image.size(0), -1).std(dim=1)
# Collect mean intensity values for each channel
channel_means_hom = {0: [], 1: [], 2: []}
channel_std_hom = {0: [], 1: [], 2: []}

for image in homogen:
    filtered_image = non_zero_pixels(image[0])
    means = mean_intensity_values(filtered_image)
    stds = standard_deviation_values(filtered_image)
    for i, mean in enumerate(means):
        channel_means_hom[i].append(mean.item())  # Convert tensor to Python scalar
    for i, std in enumerate(stds):
        channel_std_hom[i].append(std.item())  # Convert tensor to Python scalar
# Create boxplots
channel_means_het = {0: [], 1: [], 2: []}
channel_std_het = {0: [], 1: [], 2: []}
for image in heterogen:
    filtered_image = non_zero_pixels(image[0])
    means = mean_intensity_values(filtered_image)
    stds = standard_deviation_values(filtered_image)
    for i, mean in enumerate(means):
        channel_means_het[i].append(mean.item())  # Convert tensor to Python scalar
    for i, std in enumerate(stds):
        channel_std_het[i].append(std.item())  # Convert tensor to Python scalar
# Create boxplots
plt.boxplot([channel_means_hom[0], channel_means_hom[1], channel_means_hom[2], channel_means_het[0], channel_means_het[1], channel_means_het[2]], labels=['homogen 1', 'homogen 2', 'homogen 3', 'heterogen 1', 'heterogen 2', 'heterogen 3'])
plt.title('HOMGEN and HETEROGEN Boxplot of Mean Intensity Values for Each Channel')
plt.ylabel('Mean Intensity')
plt.show()

plt.boxplot([channel_std_hom[0], channel_std_hom[1], channel_std_hom[2], channel_std_het[0], channel_std_het[1], channel_std_het[2]], labels=['homogen 1', 'homogen 2', 'homogen 3', 'heterogen 1', 'heterogen 2', 'heterogen 3'])
plt.title('HOMGEN and HETEROGEN Boxplot of std Values for Each Channel')
plt.ylabel('standart deviation')
plt.show()
```
