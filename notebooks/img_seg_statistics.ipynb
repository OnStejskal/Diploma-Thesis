```python
from PIL import Image
import numpy as np
import os
import pandas as pd
from torchvision.transforms import (
    Compose,
    RandomHorizontalFlip,
    RandomVerticalFlip,
    Resize,
    ToTensor,
)
import matplotlib.pyplot as plt
def compute_std_of_green_segments(image_path, segmentation_path):
    # Load the original image and the segmentation mask
    image = Image.open(image_path)
    segmentation = Image.open(segmentation_path)
    TRANSFORMATIONS_TORCH = Compose(
    [
        Resize((256, 256)),
    ])
    # print(image.shape)
    image = TRANSFORMATIONS_TORCH(image)
    # Convert them to NumPy arrays
    image_np = np.array(image)
    segmentation_np = np.array(segmentation)

    # Define the green color in the segmentation mask
    # Assuming green is represented as [0, 255, 0]
    green_color = [0, 255, 0]

    # Create a mask for green segments
    green_mask = np.all(segmentation_np == green_color, axis=-1)

    # Extract pixels from the original image where the mask is True
    green_pixels = image_np[green_mask]

    # Compute the standard deviation of these pixels
    std_green_pixels = np.std(green_pixels, axis=0)

    return std_green_pixels

def compute_mean_of_green_segments(image_path, segmentation_path):
    # Load the original image and the segmentation mask
    image = Image.open(image_path)
    segmentation = Image.open(segmentation_path)
    TRANSFORMATIONS_TORCH = Compose(
    [
        Resize((256, 256)),
    ])
    # print(image.shape)
    image = TRANSFORMATIONS_TORCH(image)
  
    # Convert them to NumPy arrays
    image_np = np.array(image)
    # print(image_np)
    # print(image_np.shape)
    segmentation_np = np.array(segmentation)

    # Define the green color in the segmentation mask
    # Assuming green is represented as [0, 255, 0]
    green_color = [0, 255, 0]

    # Create a mask for green segments
    green_mask = np.all(segmentation_np == green_color, axis=-1)

    # Extract pixels from the original image where the mask is True
    green_pixels = image_np[green_mask]

    # Compute the mean of these pixels
    mean_green_pixels = np.mean(green_pixels, axis=0)

    return mean_green_pixels

# Usage

img_path = os.path.join("data","carotid_key_inpainted_dataset", "MAAG_FIX", "localized")
seg_path = os.path.join("data","carotid_key_inpainted_dataset","MAAG_FIX",  "segmentations")

class1_mean = []
class2_mean = []
class3_mean = []
class4_mean = []
class1_std = []
class2_std = []
class3_std = []
class4_std = []
df = pd.read_csv(os.path.join("parameters_computation", "data", "forecast_key_dataset.csv"))
for img_name in os.listdir(seg_path):
    # print(img_name)
    label_value = df.loc[df['img_name'] == img_name, 'echogenicity']
    # print(label_value)
    # print(label_value.values[0])
    if len(label_value.values) ==0:
        continue
    # print(label_value.values[0])
    label_value = label_value.values[0]
    # print(img_name, label_value)
    image_path = os.path.join(img_path, img_name)  # Replace with the path to your image
    segmentation_path = os.path.join(seg_path, img_name) 
    mean = compute_mean_of_green_segments(image_path, segmentation_path)
    std = compute_std_of_green_segments(image_path, segmentation_path)
    # print("Mean of green segmented pixels:", mean)
    mean = np.mean(mean)
    std = np.mean(std)
    if  np.isnan(mean).any() or np.isnan(std):
        continue
    if label_value == 1.:
        class1_mean.append(mean)
        class1_std.append(mean)
    elif label_value == 2.:
        class2_mean.append(mean)
        class2_std.append(mean)
    elif label_value == 3.:
        class3_mean.append(mean)
        class3_std.append(mean)
    elif label_value == 4.:
        class4_mean.append(mean)
        class4_std.append(mean)
    else:
        print(label_value)
        print("ups")

import matplotlib.pyplot as plt

# Data for the two boxplots


# Combine the data into a list
combined_data_mean = [class1_mean, class2_mean,class3_mean, class4_mean]

# Create a figure and axis
fig, ax = plt.subplots()

# Create the boxplots side by side
ax.boxplot(combined_data_mean)

# Add labels to the x-axis
ax.set_xticklabels(['1', '2', '3', '4'])

# Add a title and labels
# plt.title('Boxplots of mean pixel intensities for predicted plaques')
plt.ylabel('Intesnity mean')

# Show the plot
plt.show()


combined_data_std = [class1_std, class2_std, class3_std, class4_std]

# Create a figure and axis
fig, ax = plt.subplots()

# Create the boxplots side by side
ax.boxplot(combined_data_std)

# Add labels to the x-axis
ax.set_xticklabels(['1', '2', '3', '4'])

# Add a title and labels
# plt.title('Two Boxplots in One Graph')
plt.ylabel('Intensity standart deviation')

# Show the plot
plt.show()
```
